### Creamos un bucle while para poder ponerlo en un archivo excel

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

# Creamos un nuevo libro de trabajo de Excel
workbook = openpyxl.Workbook()

# Seleccionamos la hoja activa
hoja_activa = workbook.active

# Establecemos la pagina base

pagina = 1

# Bucle
while True:
    url = f"https://simple.ripley.com.pe/tecnologia/computacion/computadoras-de-escritorio?s=mdco&page={pagina}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Extraemos los nombres de los productos
    nombres = []
    elementos_nombres = soup.find_all("div", class_="catalog-product-details__name")
    for elemento_n in elementos_nombres:
        nombres.append(elemento_n.text.strip())

    # Extraemos los precios de los productos
    precios = []
    elementos_precios = soup.find_all("li", title="Precio Internet")
    for elemento_p in elementos_precios:
        precios.append(elemento_p.text.strip())

    # Verificamos si hay productos en la página actual
    if len(nombres) == 0:
        break

    # Imprimimos los nombres y precios de los productos
    for nombre, precio in zip(nombres, precios):
        print(nombre, precio)
        
        hoja_activa.append([nombre, precio])   # Agregamos los datos a la hoja activa del archivo Excel

    # Incrementar el número de página para la siguiente iteración
    pagina += 1
```

###  Guardamos el archivo de Excel

```python
workbook.save("precios5.xlsx")
```

### Creremos un programa que nos recomiende productos deacuerdo al precio que introducimos

```python
# Importamos librerias

import pandas as pd
import numpy as np
```

```python
# Abrimos el archivo excel
datos = pd.read_excel("precios5.xlsx",header=None)
datos
```

###  Revisamos los datos y agregaremos nombres a las columnas

```python
datos.info()
```

```python
# Creamos una nueva tabla con los nombres de las columnas
nuevos_datos = datos.rename(columns={0: "Nombre", 1: "Precio"})
nuevos_datos
```

```python
nuevos_datos.info()
```

### Convertiremos los datos no numericos 

```python
# Usaremos la libreria regular expressión para eliminar todos los caracteres que no son dígitos en la columna "Precio".
import re

nuevos_datos_modificados = nuevos_datos.copy()
nuevos_datos_modificados['Precio'] = nuevos_datos_modificados['Precio'].apply(lambda x: re.sub('[^\d]+', '', x))

# Imprimir el nuevo DataFrame
nuevos_datos_modificados
```

```python
nuevos_datos_modificados.info()
```

### Converción de valores

```python
# Convertimos la columna "Precios" en valores numéricos
nuevos_datos_modificados['Precio'] = pd.to_numeric(nuevos_datos_modificados['Precio'])
nuevos_datos_modificados.info()
```

### Programa de recomendación de precios

```python
# Ingrese un numero
print(" -------------Bienvenid@ a la tienda Ripley Perú----------------")
print("\n")
print("En esta plataforma podra encontrar un producto  adecuado asu capital disponible, disfrute su estancia. ")
print("\n")
numero = float(input("Por favor, ingrese su capital disponible: "))
print("\n")

nuevos_datos_modificados["Diferencia"] = abs(nuevos_datos_modificados["Precio"] - numero)
nuevos_datos_modificados = nuevos_datos_modificados.sort_values("Diferencia")

# Obtener los precios más cercanos al capital ingresado
precios_cercanos = nuevos_datos_modificados.loc[nuevos_datos_modificados["Diferencia"] == nuevos_datos_modificados['Diferencia'].min()]

print("Productos recomendados de acuerdo a su capital:")
print("\n")
print(precios_cercanos)
```
